```python
arr = [5,10]
s = 20


dp=[False]*(s+1)

dp[0]=True

for num in arr:
    for i in range(s,num-1,-1):
        dp[i]=dp[i] or dp[i-num]


#for i in range(s+1):
    #print(i,dp[i])
    
    
dp1=[[] for _ in range(s+1)]

    
for i in range(s,-1,-1):
    if not dp[i]:
        continue
    if dp[i]:
        subset=[]
        rem=i
        used=[False]*len(arr)
        
        for idx,num in enumerate(arr):
            if not used[idx] and rem-num>=0 and dp[rem-num]:
                subset.append(num)
                used[idx]=True
                rem-=num
            if rem==0:
                break

        dp1[i]=subset if rem==0 else []
        
for i in range(s+1):
    print(i,dp1[i])
    
```

```python
def print_dp_table(dp, weights, values, W):
    n = len(weights)

    print("\nDP Table (rows = items, columns = capacity)\n")

    # Print capacity header
    print("      W→", end="")
    for w in range(W + 1):
        print(f"{w:4}", end="")
    print("\n" + "-" * (6 + 4 * (W + 1)))

    # Print each row
    for i in range(n + 1):
        if i == 0:
            print("i=0   ", end="")   # No items
        else:
            print(f"i={i}   ", end="")  # i items considered

        for w in range(W + 1):
            print(f"{dp[i][w]:4}", end="")
        print()
```

```python
def knapsack(weights, values, W):
    n = len(weights)
    dp = [[0]*(W+1) for _ in range(n+1)]

    for i in range(1, n+1):  #weights and values
        for w in range(W+1): #capacity
            if weights[i-1] <= w:
                dp[i][w] = max(
                    values[i-1] + dp[i-1][w-weights[i-1]],
                    dp[i-1][w]
                )
            else:
                dp[i][w] = dp[i-1][w]

    print_dp_table(dp, weights, values, W)
    return dp[n][W]


# -------- Test Case --------
weights = [2, 3, 4, 5]
values  = [3, 4, 5, 6]
W = 5

result = knapsack(weights, values, W)
print("Maximum value in knapsack:", result)
```
