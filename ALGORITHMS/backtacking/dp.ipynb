```python
arr = [5,10]
s = 20


dp=[False]*(s+1)

dp[0]=True

for num in arr:
    for i in range(s,num-1,-1):
        dp[i]=dp[i] or dp[i-num]


#for i in range(s+1):
    #print(i,dp[i])
    
    
dp1=[[] for _ in range(s+1)]

    
for i in range(s,-1,-1):
    if not dp[i]:
        continue
    if dp[i]:
        subset=[]
        rem=i
        used=[False]*len(arr)
        
        for idx,num in enumerate(arr):
            if not used[idx] and rem-num>=0 and dp[rem-num]:
                subset.append(num)
                used[idx]=True
                rem-=num
            if rem==0:
                break

        dp1[i]=subset if rem==0 else []
        
for i in range(s+1):
    print(i,dp1[i])
    
```

```python
def print_dp_table(dp, weights, values, W):
    n = len(weights)

    print("\nDP Table (rows = items, columns = capacity)\n")

    # Print capacity header
    print("      W→", end="")
    for w in range(W + 1):
        print(f"{w:4}", end="")
    print("\n" + "-" * (6 + 4 * (W + 1)))

    # Print each row
    for i in range(n + 1):
        if i == 0:
            print("i=0   ", end="")   # No items
        else:
            print(f"i={i}   ", end="")  # i items considered

        for w in range(W + 1):
            print(f"{dp[i][w]:4}", end="")
        print()
```

```python
def knapsack(weights, values, W):
    n = len(weights) # no of items available
    dp = [[0]*(W+1) for _ in range(n+1)]

    for i in range(1, n+1):  #weights and values
        for w in range(W+1): #capacity
            if weights[i-1] <= w:#current weight
                dp[i][w] = max(   #max value of first i items with capacity w
                    values[i-1] + dp[i-1][w-weights[i-1]],
                    dp[i-1][w]
                )
            else:
                dp[i][w] = dp[i-1][w]

    print_dp_table(dp, weights, values, W)
    return dp[n][W]



result = knapsack(weights, values, W)
print("Maximum value in knapsack:", result)
```

```python
def gridtraveller(m,n):
    dp=[[0 for _ in range(n+1)] for _ in range(m+1)]
    dp[1][1]=1
    for i in range(m+1):
        for j in range(n+1):
            current=dp[i][j]
            if j+1<=n:
                dp[i][j+1]+=current
            if i+1<=m:
                dp[i+1][j]+=current


    for row in dp:
        print(row)
    return dp[m][n]

r=gridtraveller(4,4)

print(r)
```

```python
class Solution:
    def generate(self, numRows: int):
        triangle = []

        for i in range(numRows):
            row = [1] * (i + 1)

            for j in range(1, i):
                row[j] = triangle[i-1][j-1] + triangle[i-1][j]

            triangle.append(row)

        return triangle
```
