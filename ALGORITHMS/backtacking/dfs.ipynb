```python

class Solution(object):
    def canVisitAllRooms(self, rooms):
        """
        :type rooms: List[List[int]]
        :rtype: bool
        """
        n=len(rooms)

        visited=[False]*(n)



        def dfs(key):
            if not visited[key]:
                visited[key]=True
                keysofroom=rooms[key]
                for i in keysofroom:
                    dfs(i)
            return

        dfs(0)

        if False in visited:
            return False
        else:
            return True
                    
```
