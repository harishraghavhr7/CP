```python
buildings = [[2,9,10],[3,7,15],[5,12,12],[15,20,10],[19,24,8]]

def merge(left,right):
    i=j=0
    h1=h2=0
    prev=0
    res=[]
    
    while i<len(left) and j<len(right):
        if left[i][0]<right[j][0]:
            x=left[i][0]
            h1=left[i][1]
            i+=1
        elif left[i][0]>right[j][0]:
            x=right[i][0]
            h2=right[j][1]
            j+=1
        else:
            x=left[i][0]
            h1=left[i][0]
            h2=right[i][0]
            i+=1
            j+=1
        curr=max(h1,h2)
        if curr!=prev:
            res.append([x,curr])
            prev=curr
        
        
    for k in range(i,len(left)):
        if left[k][1]!=prev:
            res.append(left[k])
            prev=left[k][1]
    
    for k in range(j,len(right)):
        if right[k][1]!=prev:
            res.append(right[k])
            prev=right[k][1]
            
    return res



def solve(buildings):
    if len(buildings)==1:
        l,r,h=buildings[0]
        return [[l,h],[r,0]]
        
    mid=len(buildings)//2
    
    leftsky=solve(buildings[:mid])
    rightsky=solve(buildings[mid:])
    
    
    return merge(leftsky,rightsky)
print(solve(buildings))
```
